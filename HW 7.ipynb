```r
example <- data.frame(
    id = c(1, 2, 3, 4, 5), 
    name = c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
    salary = c(623.30, 515.20, 611.00, 729.00, 843.25)
)
example
```

```r
example_update <- cbind(example, department = c("accounting", "human resources", "customer service", 
"sales", "technical support"  ))
example_update
```

```r
extracted <- example_update[c(1,3,5),2:3]
extracted
```

```r
y <- c(extracted[,2])
barplot(y, names.arg = extracted$name, xlab = "Employee", ylab = "Salary")
```

```r
values <- c(max(y),min(y),median(y))
mylabel <- c("Max Salary", "Min Salary", "Median Salary")
```

```r
values
```

```r
pie(values,label = mylabel, main = "Salary Breakdown")
```
